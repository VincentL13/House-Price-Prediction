# Data Load and First Look
Load train dataset and see basic information, get an initial understanding with the meta data.

Inspect the attributes with null value, decide manuplation strategies:
- Delete attributes
- Fullfill with mean
- Fullfill with mode
- Fullfill with median


```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/House Price prediction/train.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

Attributes needed to be inspected:

LotFrontage, Alley, MasVnrType, MasVnrArea, BsmtQual, BsmtCond, BsmtExposure,
BsmtFinType1, BsmtFinType2, Electrical, FireplaceQu, GarageType
GarageYrBlt, GarageFinish, GarageQual, GarageCond,
PoolQC, Fence, MiscFeature

```python
df['LotFrontage'].describe()
# replace by median because the distribution is wild
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
df['LotFrontage'].describe()
```

```python
df['Alley'].describe()
df['Alley'].value_counts()
# fill the null by 0 to represent 'No alley access' because it is meaningful for NA
```

```python
df['Alley'] = df['Alley'].fillna(0)
df['Alley'].value_counts()
```

```python
df['Alley'].replace({0 : 'No alley access'}, inplace=True)
df['Alley'].value_counts()
```

```python
# check the df info to ensure it be modified correctly
df.info()
```

```python
# For MasVnrType
# We just remove the attribute
df_1 = df.drop(columns=['MasVnrType'])
df_1.info()
```

```python
# MasVnrArea
df_1['MasVnrArea'].describe()
df_1['MasVnrArea'].value_counts()
```

```python
# Fill the NA with 0
df_1['MasVnrArea'] = df_1['MasVnrArea'].fillna(0)
df_1['MasVnrArea'].describe()
```

```python
df_1['BsmtQual'].value_counts()
```

```python
# Replace the NA by 'no basement'
df_1['BsmtQual'] = df_1['BsmtQual'].fillna('No Basement')
df_1['BsmtQual'].value_counts()
df_1['BsmtQual'].info()
```

```python
df_1.to_csv('/content/drive/MyDrive/Colab Notebooks/House Price prediction/df_1.csv', index=False)
```
